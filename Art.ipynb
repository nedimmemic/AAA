```python
message_david = """
DAVID

Michelangelo's David: A Symbol of Strength and Perfection

Introduction
Michelangelo's David, a masterpiece of Renaissance sculpture, stands as one of the most iconic representations of human beauty and strength. Carved between 1501 and 1504, this marble statue depicts the Biblical hero David, who defeated the giant Goliath with a single stone.
David is currently housed in the Galleria dell'Accademia in Florence, Italy.

Description
Michelangelo's David is a colossal statue, standing at an impressive 17 feet (5.17 meters) tall. It portrays a nude David in a contrapposto stance, a relaxed yet alert pose that conveys both his physical power and psychological readiness.
David's gaze is focused and intense, symbolizing his contemplation and courage before the battle.

Artistic Techniques
Michelangelo's David showcases several remarkable artistic techniques:
- Anatomical Precision: Michelangelo's deep understanding of human anatomy is evident in the detailed rendering of muscles, veins, and tendons. Every part of David's body is meticulously sculpted, giving the statue a lifelike appearance.
- Contrapposto Stance: This classical pose, where one leg bears the weight while the other is relaxed, creates a sense of dynamic balance and naturalism. It enhances the statue's sense of movement and life.
- Proportionality: Despite its size, the proportions of David's body are harmonious and realistic, demonstrating Michelangelo's mastery of scale and form.

Commentary
David is not only a testament to Michelangelo's technical skill but also a powerful symbol of Florence's civic pride and ideals.
At the time, Florence saw itself as a "David" standing against the "Goliath" of rival states and external threats. The statue embodies the values of courage, strength, and faith in the face of adversity.

One striking feature is David's oversized hands and head, which have been interpreted in various ways.
Some suggest it emphasizes his intellectual and physical strength, while others believe it highlights his role as a contemplative and divine hero.

Legacy
Michelangelo's David remains one of the most celebrated and studied sculptures in art history. It has inspired countless artists and continues to captivate millions of visitors each year. The statue's blend of idealized beauty, anatomical precision, and emotional depth makes it a timeless symbol of human potential and artistic excellence.
"""

print(message_david)
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Updated URLs of famous works with valid links
works = {
    "Leonardo da Vinci": "https://upload.wikimedia.org/wikipedia/commons/thumb/e/ec/Mona_Lisa%2C_by_Leonardo_da_Vinci%2C_from_C2RMF_retouched.jpg/1200px-Mona_Lisa%2C_by_Leonardo_da_Vinci%2C_from_C2RMF_retouched.jpg",
    "Michelangelo": "https://th.bing.com/th/id/OIP.yjSoidPUmQD-RlwQY8xKkwHaKc?rs=1&pid=ImgDetMain",
    "Raphael": "https://upload.wikimedia.org/wikipedia/commons/thumb/7/7a/RAFAEL_-_Madonna_Sixtina_%28Gem%C3%A4ldegalerie_Alter_Meister%2C_Dresden%2C_1513-14._%C3%93leo_sobre_lienzo%2C_265_x_196_cm%29.jpg/1200px-RAFAEL_-_Madonna_Sixtina_%28Gem%C3%A4ldegalerie_Alter_Meister%2C_Dresden%2C_1513-14._%C3%93leo_sobre_lienzo%2C_265_x_196_cm%29.jpg",
    "Claude Monet": "https://www.antiquesandthearts.com/wp-content/uploads/2019/10/Claude-Monet-Waterlilies-and-Japanese-Bridge-1899_HIGH-RES.jpg",
    "Salvador Dalí": "https://upload.wikimedia.org/wikipedia/en/d/dd/The_Persistence_of_Memory.jpg",
    "Pablo Picasso": "https://upload.wikimedia.org/wikipedia/en/7/74/PicassoGuernica.jpg",
    "Caravaggio": "https://th.bing.com/th/id/OIP.aT2JN7_jF7NkmI2Ekg2EJQHaF8?w=9116&h=7321&rs=1&pid=ImgDetMain",
    "Rembrandt": "https://th.bing.com/th/id/R.1816083b72ad4eabb4f07866ee02e31b?rik=2jS%2bgRyquv3NiA&pid=ImgRaw&r=0",
    "Vermeer": "https://th.bing.com/th/id/OIP.UxdbEC4JzyMPV1QtDL-dQwAAAA?rs=1&pid=ImgDetMain"
}

# Define headers for User-Agent compliance
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Load images and resize them to the same dimensions
images = []
for artist, url in works.items():
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Check if the request was successful
        img = Image.open(BytesIO(response.content))
        img = img.resize((300, 300))  # Resize to a fixed dimension
        images.append((artist, img))
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {artist}'s work: {e}")
    except Exception as e:
        print(f"Error processing {artist}'s work: {e}")

# Create a blank canvas
combined_width = 3 * 300  # 3 images in a row, each 300 pixels wide
combined_height = 3 * 300  # 3 images in a column, each 300 pixels tall
combined_image = Image.new("RGB", (combined_width, combined_height))

# Paste images onto the canvas
for i, (artist, img) in enumerate(images):
    x = (i % 3) * 300  # Column position
    y = (i // 3) * 300  # Row position
    combined_image.paste(img, (x, y))

# Display the combined image using matplotlib
plt.figure(figsize=(12, 12))
plt.imshow(combined_image)
plt.axis('off')  # Hide the axes
plt.title("Best Works of Leonardo da Vinci, Michelangelo, Claude Monet, Raphael, Salvador Dalí, Pablo Picasso, Caravaggio, Rembrandt, and Vermeer")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for painters
painters = [
    {"name": "Leonardo da Vinci", "birth_year": 1452, "death_year": 1519},
    {"name": "Michelangelo", "birth_year": 1475, "death_year": 1564},
    {"name": "Raphael", "birth_year": 1483, "death_year": 1520},
    {"name": "Rembrandt", "birth_year": 1606, "death_year": 1669},
    {"name": "Vincent van Gogh", "birth_year": 1853, "death_year": 1890},
    {"name": "Pablo Picasso", "birth_year": 1881, "death_year": 1973},
    {"name": "Claude Monet", "birth_year": 1840, "death_year": 1926},
    {"name": "Salvador Dalí", "birth_year": 1904, "death_year": 1989},
    {"name": "Frida Kahlo", "birth_year": 1907, "death_year": 1954},
    {"name": "Jackson Pollock", "birth_year": 1912, "death_year": 1956},
]

# Data for musicians
musicians = [
    {"name": "Johann Sebastian Bach", "birth_year": 1685, "death_year": 1750},
    {"name": "Wolfgang Amadeus Mozart", "birth_year": 1756, "death_year": 1791},
    {"name": "Ludwig van Beethoven", "birth_year": 1770, "death_year": 1827},
    {"name": "Frédéric Chopin", "birth_year": 1810, "death_year": 1849},
    {"name": "Pyotr Ilyich Tchaikovsky", "birth_year": 1840, "death_year": 1893},
    {"name": "Claude Debussy", "birth_year": 1862, "death_year": 1918},
    {"name": "Igor Stravinsky", "birth_year": 1882, "death_year": 1971},
    {"name": "Dmitri Shostakovich", "birth_year": 1906, "death_year": 1975},
    {"name": "John Lennon", "birth_year": 1940, "death_year": 1980},
    {"name": "David Bowie", "birth_year": 1947, "death_year": 2016},
]

# Data for writers
writers = [
    {"name": "William Shakespeare", "birth_year": 1564, "death_year": 1616},
    {"name": "Jane Austen", "birth_year": 1775, "death_year": 1817},
    {"name": "Charles Dickens", "birth_year": 1812, "death_year": 1870},
    {"name": "Leo Tolstoy", "birth_year": 1828, "death_year": 1910},
    {"name": "Mark Twain", "birth_year": 1835, "death_year": 1910},
    {"name": "Virginia Woolf", "birth_year": 1882, "death_year": 1941},
    {"name": "James Joyce", "birth_year": 1882, "death_year": 1941},
    {"name": "Franz Kafka", "birth_year": 1883, "death_year": 1924},
    {"name": "George Orwell", "birth_year": 1903, "death_year": 1950},
    {"name": "Gabriel García Márquez", "birth_year": 1927, "death_year": 2014},
]

# Extracting data for plotting
def extract_data(category):
    names = [c["name"] for c in category]
    birth_years = [c["birth_year"] for c in category]
    death_years = [c["death_year"] for c in category]
    lifespans = [c["death_year"] - c["birth_year"] for c in category]
    return names, birth_years, death_years, lifespans

names_painters, birth_years_painters, death_years_painters, lifespans_painters = extract_data(painters)
names_musicians, birth_years_musicians, death_years_musicians, lifespans_musicians = extract_data(musicians)
names_writers, birth_years_writers, death_years_writers, lifespans_writers = extract_data(writers)

# Plotting the data
plt.figure(figsize=(14, 8))
plt.barh(names_painters, lifespans_painters, left=birth_years_painters, color='gold', label='Painters')
plt.barh(names_musicians, lifespans_musicians, left=birth_years_musicians, color='lightgreen', label='Musicians')
plt.barh(names_writers, lifespans_writers, left=birth_years_writers, color='orchid', label='Writers')
plt.xlabel('Year')
plt.ylabel('Individuals')
plt.title('Lifespans of Renowned Painters, Musicians, and Writers')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Annotate the birth and death years
categories = [painters, musicians, writers]
colors = ['gold', 'lightgreen', 'orchid']

for cat_index, category in enumerate(categories):
    names, birth_years, death_years, _ = extract_data(category)
    for i in range(len(names)):
        plt.text(birth_years[i], sum(len(extract_data(c)[0]) for c in categories[:cat_index]) + i, f"{birth_years[i]}", va='center', ha='right', fontsize=8, color='black')
        plt.text(death_years[i], sum(len(extract_data(c)[0]) for c in categories[:cat_index]) + i, f"{death_years[i]}", va='center', ha='left', fontsize=8, color='black')

plt.show()
```
